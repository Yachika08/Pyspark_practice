```python
#Creating dataframe
data = [("Aman",'IT',30000,'New Delhi',2000), ("Abhi", 'UI',35000,'Haryana',1200),\
 ("Annu", 'IT',35000,'Punjab',1200),("Akshay", 'IT',25000,'New Delhi',1500),("Abhi", 'UI',40000,'Haryana',1000)]
df = spark.createDataFrame(data, ["Name", "department","salary","State","Incentive"])
print("Original Dataframe")
df.show()
```

```python
#Group by on multiple columns
grouped_df = df.groupby("department","state").sum("salary","Incentive")
print('After Groupby on the department and state')
grouped_df.show()

print("Adding name columnn in group by")
df.groupby("Name","department","state").sum("salary","Incentive").show()
```

```python
#Aliasing in groupby
from pyspark.sql.functions import sum,avg,max
df.groupby("department")\
  .agg(sum("salary").alias("Sum_salary"),\
     avg("salary").alias("Avg_salary"),\
     sum("Incentive").alias("Sum_bonus"),\
     max("Incentive").alias("Max_bonus"),\
).show()
```

```python
#filter condition in group by
print("with filer condition where bonus is greater than 2500")
df.groupby("department")\
  .agg(sum("salary").alias("Sum_salary"),\
     avg("salary").alias("Avg_salary"),\
     sum("Incentive").alias("Sum_bonus"),\
     max("Incentive").alias("Max_bonus"))\
    .where(col("sum_bonus") >= 2500).show()
```

```python
#filter condition in group by
print("with filer condition where bonus is less than 2500")
df.groupby("department")\
  .agg(sum("salary").alias("Sum_salary"),\
     avg("salary").alias("Avg_salary"),\
     sum("Incentive").alias("Sum_bonus"),\
     max("Incentive").alias("Max_bonus"))\
    .where(col("sum_bonus") <= 2500).show()
```

```python
from pyspark.sql.functions import col, initcap, upper, lower, substring, lpad, rpad, ltrim, rtrim
# Sample data
data = [("John Doe",),
        ("Jane Smith",),
        ("Adam Johnson",)]

# Create the DataFrame
df = spark.createDataFrame(data, ["name"])
```

```python
#String Functions
print("Output for contains")
df_contains = df.filter(col('name').contains('ohn'))
df_contains.show()
print("Output for startswith")
df_startswith = df.filter(col('name').startswith('J'))
df_startswith.show()
print("Output for endswith")
df_endswith = df.filter(col('name').endswith('Smith'))
df_endswith.show()
print("Output for initcap")
df_initcap = df.select(initcap(df.name).alias('initcap_name'))
df_initcap.show()
print("Output for upper")
df_upper = df.select(upper(df.name).alias('upper_name'))
df_upper.show()
print("Output for lower")
df_lower = df.select(lower(df.name).alias('lower_name'))
df_lower.show()
print("Output for substring")
df_substring = df.select(substring('name', 2, 4).alias('substring_name'))
df_substring.show()
print("Output for lpad")
df_lpad = df.select(lpad('name', 12, 'X').alias('lpad_name'))
df_lpad.show()
print("Output for rpad")
df_rpad = df.select(rpad(df.name, 12, 'Y').alias('rpad_name'))
df_rpad.show()
print("Output for ltrim")
df_ltrim = df.select(ltrim(df.name).alias('ltrim_name'))
df_ltrim.show()
print("Output for rtrim")
df_rtrim = df.select(rtrim(df.name).alias('rtrim_name'))
df_rtrim.show()
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number, rank, dense_rank, min, max, sum, avg, countDistinct,lag,lead

Data = (("Kirti", "Sales", 3000), \
    ("Leo", "Sales", 4600),  \
    ("Neha", "Sales", 4100),   \
    ("Priya", "Finance", 3000),  \
    ("Pooja", "Sales", 3000),    \
    ("Anjali", "Finance", 3300),  \
    ("Kritika", "Finance", 3900),    \
    ("Megha", "Marketing", 3000), \
    ("Megha", "Marketing", 2000),\
    ("Shweta", "Sales", 4100) \
  )

columns= ["employee_name", "department", "salary"]

df = spark.createDataFrame(data = Data, schema = columns)

print("ORIGINAL DF")
df.show()

#Create a window specification
windowspec = Window.partitionBy("department").orderBy("salary")
print("using window function with orderby")
#calcualte row number
#returns sequential number starting from 1 within window partition
print("row number")
df.withColumn("row_number",row_number().over(windowspec)).show()
#calculate rank
print("Rank")
df.withColumn("rank",rank().over(windowspec)).show()
#dense rank
print("Dense Rank")
df.withColumn("dense_rank",dense_rank().over(windowspec)).show()
#Calculate sum
print("SUM COLUMN")
df.withColumn("sum_column",sum("salary").over(windowspec)).show()
#Calculate min
print("MIN COLUMN")
df.withColumn("min_column",min("salary").over(windowspec)).show()
#Calculate max
print("MAX COLUMN")
df.withColumn("max_column",max("salary").over(windowspec)).show()
#Calculate avg
print("AVG COLUMN")
df.withColumn("avg_column",avg("salary").over(windowspec)).show()
#calculate LAG
print("LAG")
df.withColumn("lag",lag("salary",1,default=0).over(windowspec)).show()
#calculate LEAD
print("LEAD")
df.withColumn("lead",lead("salary",1,default=0).over(windowspec)).show()
```
